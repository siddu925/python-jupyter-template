### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the line chart
x_line = np.arange(0, 5)
y_line = x_line * 2

# Data for the bar chart
categories_bar = ['X', 'Y', 'Z']
values_bar = [10, 15, 7]

# Data for the scatter plot
x_scatter = [1, 2, 3, 4]
y_scatter = [3, 1, 4, 2]

# Data for the pie chart
labels_pie = ['Part 1', 'Part 2', 'Part 3']
sizes_pie = [30, 45, 25]
colors_pie = ['gold', 'silver', 'lightcoral']
explode_pie = (0.1, 0, 0)  # explode 1st slice

# Create the subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('2x2 Subplot Layout')

# First subplot (line chart)
axs[0, 0].plot(x_line, y_line, marker='o')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].grid(True)

# Second subplot (bar chart)
axs[0, 1].bar(categories_bar, values_bar, color='skyblue')
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Third subplot (scatter plot)
axs[1, 0].scatter(x_scatter, y_scatter, color='green')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')
axs[1, 0].grid(True)

# Fourth subplot (pie chart)
axs[1, 1].pie(sizes_pie, explode=explode_pie, labels=labels_pie, colors=colors_pie, autopct='%1.1f%%', shadow=True, startangle=90)
axs[1, 1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axs[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust for suptitle

# Show the plots
plt.show()

# It's possible you had
```
