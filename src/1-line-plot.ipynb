### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.arange(0, 11)

# Calculate y values (square of x)
y = x ** 2

# Create the plot
plt.plot(x, y)

# Add title
plt.title('Line Plot of y = x^2')

# Add x-axis label
plt.xlabel('x')

# Add y-axis label
plt.ylabel('y = x^2')

# Add grid
plt.grid(True)

# Show the plot
plt.show()
```
